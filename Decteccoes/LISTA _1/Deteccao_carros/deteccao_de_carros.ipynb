```python
import cv2
from PIL import Image
from IPython.display import display
```

```python
imagem = cv2.imread("car.jpg")
rgb_img = cv2.cvtColor(imagem,cv2.COLOR_BGR2RGB)
detector_carros = cv2.CascadeClassifier('cars.xml') 
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

deteccoes_carros = detector_carros.detectMultiScale(imagem_cinza, scaleFactor=1.06, minNeighbors=2, minSize=(25,25))
for (x, y, w, h) in deteccoes_carros:
    cv2.rectangle(rgb_img, (x, y), (x + w, y + h), (0,0, 255), 2)

display(Image.fromarray(rgb_img))
```
