```python
import cv2
from PIL import Image
from IPython.display import display
```

```python
imagem = cv2.imread("people3.jpg")
rgb_img = cv2.cvtColor(imagem,cv2.COLOR_BGR2RGB)
detector_fullbody = cv2.CascadeClassifier('fullbody.xml') 
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

deteccoes_fullbody = detector_fullbody.detectMultiScale(imagem_cinza, scaleFactor=1.05, minNeighbors=5, minSize=(60,60))
for (x, y, w, h) in deteccoes_fullbody:
    cv2.rectangle(rgb_img, (x, y), (x + w, y + h), (0,0, 255), 2)

display(Image.fromarray(rgb_img))
```
