```python
import cv2
from PIL import Image
from IPython.display import display
```

```python
imagem = cv2.imread("clock.jpg")
rgb_img = cv2.cvtColor(imagem,cv2.COLOR_BGR2RGB)
detector_clock = cv2.CascadeClassifier('clocks.xml') 
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

deteccoes_clock = detector_clock.detectMultiScale(imagem_cinza, scaleFactor=1.1, minNeighbors= 1)
for (x, y, w, h) in deteccoes_clock:
    cv2.rectangle(rgb_img, (x, y), (x + w, y + h), (0,0, 255), 2)

display(Image.fromarray(rgb_img))
```
