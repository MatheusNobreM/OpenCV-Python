```python
import cv2
from PIL import Image
from IPython.display import display
```

```python
imagem = cv2.imread("people1.jpg")
rgb_img = cv2.cvtColor(imagem,cv2.COLOR_BGR2RGB)
```

```python
detector_olhos = cv2.CascadeClassifier('haarcascade_eye.xml') 
```

```python
#imagem = cv2.resize(imagem,(800, 600))
#rgb_img = cv2.resize(rgb_img,(800, 600))
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

#detector_facial =cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
#deteccoes = detector_facial.detectMultiScale(imagem_cinza, scaleFactor= 1.09, minSize=(32,32))
#for (x, y, w, h) in deteccoes:
#   cv2.rectangle(rgb_img, (x, y), (x + w, y + h), (0,255, 255), 2)

deteccoes_olhos = detector_olhos.detectMultiScale(imagem_cinza, scaleFactor= 1.09, minNeighbors=13, maxSize=(50,50))
for (x, y, w, h) in deteccoes_olhos:
   cv2.rectangle(rgb_img, (x, y), (x + w, y + h), (0,0, 255), 2)

display(Image.fromarray(rgb_img))
```
