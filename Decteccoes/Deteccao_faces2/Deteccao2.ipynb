```python
import cv2
from PIL import Image
from IPython.display import display
```

```python
imagem = cv2.imread("people2.jpg")
rgb_img = cv2.cvtColor(imagem,cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_img))
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
detector_facial =cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
deteccoes = detector_facial.detectMultiScale(imagem_cinza, scaleFactor= 1.2, minNeighbors= 3,
                                            minSize=(32,32), maxSize=(100,100))
for x, y, w, h in deteccoes:
  print(w,h)
  cv2.rectangle(rgb_img, (x, y), (x + w, y + h), (0,255, 255), 2)
display(Image.fromarray(rgb_img))
```
